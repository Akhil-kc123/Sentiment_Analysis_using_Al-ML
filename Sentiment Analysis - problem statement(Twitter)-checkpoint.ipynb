<h3> Data Cleaning </h3>

<h4> Removing the stopwords and handling unnecessory words </h4>

```python
# defining stemmer and stopword for english language
nltk.download('stopwords')
stemmer = nltk.SnowballStemmer('english')
stopword = set(stopwords.words('english'))
```

```python
#defining a function for cleaning tweets
def clean_data(text):
    text = str(text).lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+➡️', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    text = re.sub(r'[0-9]', '',text)  
    text = re.sub('¾➡️','',text)
    text = [word for word in text.split(' ') if word not in stopword]
    text = " ".join(text)
    text = [stemmer.stem(word) for word in text.split(' ')]
    text=" ".join(text)
    return text
```

```python
# cleaning tweet column by passing to above function
df['tweets'] = df['tweets'].apply(clean_data)
```

<h3> Analyzing sentiment</h3>

```python
# custom function for sentiment analyze
def sentiment_analyze(text):
    analyze = TextBlob(text)
    if analyze.sentiment.polarity > 0:
        return 1
    elif analyze.sentiment.polarity == 0:
        return 0
    else:
        return -1
```

```python
# applying above fn and creating new column
df['Sentiment_analysis'] = np.array([sentiment_analyze(tweet) for tweet	in df['tweets']])
```

```python
# count plot for sentiment analysis
sns.countplot(x='Sentiment_analysis', data=df)
plt.show()
```

<h3> Data splitting </h3>

```python
# data splitting
X = df['tweets']
Y = df['Sentiment_analysis']
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
```

<h3> Data Transformation usiing TF-IDf Vectorizer </h3>

```python
# transforming data
tfidf = TfidfVectorizer(max_features=6000)
X_train_tfidf = tfidf.fit_transform(x_train)
X_test_tfidf = tfidf.transform(x_test)
X_train_tfidf
```

<h3> Model building and Evaluation</h3>

```python
# creating logistic regression model
model = LogisticRegression()
model.fit(X_train_tfidf, y_train)
```

```python
# defining fn for evaluation
def evaluate_model(model, X_test, y_test):
    y_pred = model.predict(X_test)
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("Classification Report:", classification_report(y_test, y_pred))

# Evaluate the model
evaluate_model(model, X_test_tfidf, y_test)    
```

<h3> Analyzing the sentiment </h3>

```python
# creating list for positive,neutral and negative sentiment tweets
positive_tweets	=	[	tweet	for	index,	tweet	in	enumerate(df['tweets'])	if	df['Sentiment_analysis'][index]	>	0] 
neutral_tweets	=	[	tweet	for	index,	tweet	in	enumerate(df['tweets'])	if	df['Sentiment_analysis'][index]	==	0] 
negitive_tweets	=	[	tweet	for	index,	tweet	in	enumerate(df['tweets'])	if	df['Sentiment_analysis'][index]	<	0]
```
